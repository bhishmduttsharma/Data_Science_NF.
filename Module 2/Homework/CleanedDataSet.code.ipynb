#Step 3: Identify and Handle Missing Values

```python
clean_df.isnull()                                                                           # TO CEHCK MISSING VALUES.
```

```python
clean_df.isnull().sum()                                                                   # TO CHECK COUNT OF MISSING VALUES.
```

```python
clean_df['Years of Experience'] = clean_df['Years of Experience'].fillna(           # FILL NUMERICAL COLUMNS WITH MEDIAN.
    clean_df['Years of Experience'].median())


clean_df['Monthly Salary (INR)'] = clean_df['Monthly Salary (INR)'].fillna(
    clean_df['Monthly Salary (INR)'].median())
```

```python
categorical_cols = [                                                                 # TO FILL THE ALPHABATE VALUE WE USE CATEGORICAL WITH MODE.
    'Status',
    'Age Group',
    'Education',
    'Industry',
    'Location',
    'AI Risk'
]
for col in categorical_cols:
    clean_df[col] = clean_df[col].fillna(clean_df[col].mode()[0])
```

```python
clean_df.isnull().sum()                                                             # CHECK AGAIN AFTER THE FILL VALUES.
```

```python
clean_df                                                                                # NOW DATA IS GOOD ATER FILL NULL VALUES.
```

#Step 4: Remove Duplicate Records

```python
clean_df.duplicated()                                                                  # TO CHECK THE DUPLICATE VALUE.
```

```python
clean_df.duplicated().sum()                                                             # TO CHECK COUNT OF DUPLICATE VALUES.
```

#Step 5: Correct Inconsistent Formatting and Text Values

```python
text_columns = ['Status', 'Education', 'Industry', 'Location', 'AI Risk']               # TO REMOVE EXTRA SPACE(.str.strip())  AND FOR CONVERT FIRST LETTER CAPITAL(.str.title())

for col in text_columns:
    clean_df[col] = clean_df[col].str.strip().str.title()
```

```python
clean_df['Education'] = clean_df['Education'].replace({                                 # TO CORRECT THE WORD WE REPLACE THE WORD.
    'Phd': 'PhD'
})
```

```python
clean_df['Age Group'] = clean_df['Age Group'].str.replace('_', '-')                     # TO HANDEL THE SPACE BETWEEN NUMBER.
```

```python
clean_df
```

#Step 6: Fix Data Type Issues

```python
clean_df.dtypes                                                                          # DATE RECORDED IN OBJECT TYPE
```

```python
clean_df['Date Recorded'] = pd.to_datetime(clean_df['Date Recorded'])
```

#Step 8: Rename and Standardize Column Names

```python
clean_df.columns = (                                                                     # TO RENAME THE COLUMNS.
    clean_df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("(", "", regex=False)
      .str.replace(")", "", regex=False)
)
```
